## Lagrange Multipliers

**Lagrange Multipliers**, or **Undetermined Multipliers**, are used to find the stationary points of a function of several variables subject to one or more constraints.

Consider the problem of finding the maximum of a function $f(x_1, x_2)$ subject to a constraints relating $x_1$ and $x_2$ which is described as follows:

$$
    g(x_1, x_2) = 0
$$
A common approach to find the optimal solutions of $f(x_1, x_2)$ is to solve the constraint $g(x_1, x_2)$ to express $x_2$ as a function of $x_1$ in the form $x_2 = h(x_1)$. Then we replace $x_2$ in $f(x_1, x_2)$ to obtain $f(x_1, h(x_1))$. The maximum with respect to $x_1$ could be determined in the usual way to give the stationary point $x_1^*$, with the corresponding stationary point of $x_2$ is $x_2^* = h(x_1^*)$.

However, not all $h(x)$ can be analytical results, or in many situations, the differential forms of $\frac{\delta f(x_1, x_2)}{\delta x_1}$ are too complicated that we can only obtain approximate solutions from the numerical perspectives. To overcome such limitations, Lagrange proposed to introduce a novel approach for finding stationary points of $f(x_1, x_2)$.

First, consider a d-dimensional variable $x = (x_1, x_2, ..., x_D)^T \in (D)$. Then the constraints $g(x) = 0$ introduces a hyperplan of dimensional $d-1$ in the space $(D)$ of $x$. This hyperplan $(D)$ will intersect $f$ by a contour of $f$. We will show that the tangent of the hyperplan $(D)$ and the tangent of the contour is parallel.

First, we show that at the saddle points $x^*$ such that $x^* \in (D)$ or $g(x^*) = C$, the derivative of $f$ is perpendicular to the tangent of the contour of $f. For the ease of notation, we denote the coordinate $x = (x_1, x_2, ..., x_D)$ as a function w.r.t the step variable $t$ as:

$$
    x: \mathbb{R} \mapsto \mathbb{R}^D, t \mapsto (x_1(t), x_2(t), ..., x_D(t))
$$
where
$$
    x_i: \mathbb{R} \mapsto \mathbb{R}^D, t \mapsto x_i(t)
$$

Then the function $f$ at step $t$ is the composite function denoted as:

$$
    h \equiv f \circ x: \mathbb{R} \mapsto \mathbb{R}^D, t \mapsto (f \circ x)(t) = f(x(t)) = f(x_1(t), x_2(t), ..., x_D(t))
$$
Let $t^*$ is a scalar so that $x^* = x(t^*)$. Obviously $t$ is also a saddle of $h$. From that on, at $t = t^*$ we have:

$$
\begin{align}
    \nabla_t h(t^*) = 0 & \Leftrightarrow \sum_{i=1}^D \frac{\delta f}{\delta x_i}(x_i^*) \frac{\delta x_i}{\delta t}(t^*) = 0 \Leftrightarrow \nabla_x f(x^*)^T \cdot \nabla_t x(t^*) = 0
\end{align}
$$

We have

$$
    \nabla_x f(x^*) = \left[\frac{\delta f}{\delta x_1}, \frac{\delta f}{\delta x_2}, ..., \frac{\delta f}{\delta x_D}\right]^T
$$
and
$$
    \nabla_t x(t^*) = \left[\frac{\delta x_1}{\delta t}, \frac{\delta x_2}{\delta t}, ..., \frac{\delta x_D}{\delta t}\right]^T
$$

That is, we have $\nabla_x f \perp \nabla_t x$ where $\nabla_t x$ is the tangent belonging to the contour line and $\nabla_x f$ is the gradient of $f$. Accordingly, we have $\nabla_x f \perp (D)$. Moreover, as $x^* \in (D)$ and $x + \epsilon \in (D)$, then $g(x + \epsilon) - g(x) \in (D)$. Making the Taylor expandation of $g(x + \epsilon)$, we have:

$$
    g(x + \epsilon) \approx g(x) + \epsilon^T \nabla g(x)
$$
Because both $x$ and $x + \epsilon$ lie in the same constraints surface $(D)$, we have $g(x) = g(x + \epsilon)$ hence this results in $\epsilon^T \nabla g(x) = 0$. However, as $\epsilon \in (D)$, which results in the vector $\epsilon$ is parallel to $(D)$. From that on, we have $\nabla g(x) \perp \epsilon$ hence $\nabla g(x) \perp (D)$. Finally, as $\nabla_x g \perp (D)$ and $\nabla_x f \perp (D)$, we get $\nabla_x f \parallel \nabla_x g$.

We can see the statement $\nabla_x f \perp (D)$ from the view of Taylor expandation. Our objective is to determine the optimal $x^*$ of the function $f(x)$ which also lies in the constraints surface $(D)$. Deploying Taylor expandation of $f(x* + \epsilon)$ we have

$$
    f(x^* + \epsilon) = f(x^*) + \epsilon^T \nabla f(x^*)
$$
As $x^*$ is the optimal point of $f$, hence we have $f(x^* + \epsilon) = f(x^*)$, so $\epsilon^T \nabla f(x) = 0$ which indicates $\nabla f(x^*) \perp (D)$, which means $\nabla f(x^*) \parallel \nabla g(x^*)$, 

We have shown that $\nabla_x f \parallel \nabla_x g$, so there is a scalar $\lambda \in \mathbb{R}$ such that

$$
    \nabla f + \lambda \nabla g = 0
$$
This $\lambda$ is called the *Lagrange multiplier*.

At this time, we define the *Lagrange* function for the ease of notation:

$$
    L(x, \lambda) \equiv f(x) + \lambda g(x)
$$
By setting the partial derivatives of $L(x, \lambda)$ with respect to (w.r.t.) $x$ to zero $\frac{\delta}{\delta x} L(x, \lambda) = 0$, we obtain $\nabla f(x) + \nabla g(x) = 0$, and by letting $\frac{\delta}{\delta \lambda} L(x, \lambda) = 0$ we have $g(x) = 0$. In other words, the optimal $x^*$ and $\lambda^*$ of the Lagrange function $L(x, \lambda)$ not only maximize $f(x)$, but also satisfy the constraints $g(x) = 0$.

## Karush - Kuhn - Tucker Multipliers

To this point, the abovementions discuss the case of $g(x) = 0$. Our next step is considering the optimal solution w.r.t the constraints $g(x) \ge 0$. In this case, there are two possible solutions: (1) the stationary points lie in the surface $g(x) = 0$ and (2) those are lie in the region $g(x) > 0$. Such stationary points in the first case are called to be *active*, while those in the second case are called to be *inactive*. 

In the former case, the constraints analogy to the discussed solutions above where we have the stationary points of the Lagrange function correspond to $\lambda \neq 0$. However, the sign of the Lagrange multipliers is somewhat worthy to examine. We have already known that the optimal points for the Lagrange function lie in the boundary $g(x) = 0$. However, as we are considering the region $g(x) > 0$, this implies that $\nabla g(x) < 0$, while $\nabla f(x) > 0$ as we are moving to the maximal points. Accordingly, we have $\nabla f(x) = -\lambda \nabla g(x)$ for some $\lambda > 0$. In the later case, the function $g(x)$ plays no role, hence the stationary condition is simply $\nabla f(x) = 0$. In either case, we have $\lambda g(x) = 0$, thus the solution of maximizing the function $f(x)$ subjects to $g(x) \ge 0$ is obtained by optimizing the Lagrange function $L(x, \lambda)$ subjects the following conditions:

\begin{align}
    g(x) & \ge & 0 \\
    \lambda & \ge & 0 \\
    \lambda g(x) & = & 0
\end{align}

In the case we have a saddle point $x^*$ that satisfies the criteria $g(x*) \ge 0$, then according to [Karush - Kuhn - Tucker theory](https://en.wikipedia.org/wiki/Karush%E2%80%93Kuhn%E2%80%93Tucker_conditions), there exists $\lambda^*$  such that $(x^*, \lambda^*)$ is a saddle point of $L(x, \lambda)$. This multiplier $\lambda$ obviously satisfies the three above conditions. These conditions are also called **Karush - Kuhn - Tucker conditions**.


In the dual case where we want to minimize the function $f(x)$ w.r.t the constraints $g(x) \ge 0$, then we minimize the Lagrange function

$$
    L(x, \lambda) = f(x) - \lambda g(x)
$$
with respect to $x$, and again $\lambda \ge 0$.
