# Kernel methods

Let consider the linear regression model with the Gaussian assumption of the noise. Maximum a Posterior (MAP) implies training this method using Sum-of-Square Error with Regularization as follows:

$$
    J(w) = \frac{1}{2} \sum_{n=1}^N (w^T \phi(x_n) - t_n)^2 + \frac{\lambda}{2} w^T w
$$

Taking its derivative with respect to (w.r.t) the parameter $w$ we have:

$$
    \frac{\delta}{\delta w} J(w) = \sum_{n=1}^N (w^T \phi(x_n) - )\phi(x_n) + \lambda w
$$

Setting this partial derivative to $0$ to get:

\begin{align}
    &                   & \frac{\delta}{\delta w} J(w) & = & 0 \\
    & \Leftrightarrow   & w & = & \sum_{n=1}^N (t_n - w^T\phi(x_n)) \phi(x_n) \\
    & \Leftrightarrow   & w & = & \sum_{n=1}^N a_n \phi(x_n) = a^T \Phi 
\end{align}
where $a_n = t_n - w^T \phi(x_n)$, $a = (a_1, a_2, ..., a_n)^T$ and $\Phi = (\phi(x_1), \phi(x_2), ..., \phi(x_n))^T$. It is worthy to note that for later convenience, we ignore the scalar factor $\lambda$.

Then replace $w$ by $a^T\Phi$ in $J(w)$, we have:

\begin{align}
    J(a) & = \frac{1}{2} \sum_{n=1}^{N} \left(a^T \Phi \phi(x_n) - t_n\right)^2 + \frac{1}{2} a^T \Phi \Phi^T a \\
    & = \frac{1}{2} \sum_{n=1}^{N} \left((a^T \Phi \phi(x_n))^T(a^T \Phi \phi(x_n)) - 2 \Phi^T a \phi(x_n) + t_n^2\right) + \frac{1}{2} a^T \Phi \Phi^T a \\
    & = \frac{1}{2} \sum_{n=1}^{N} \left(\phi(x_n)^T \Phi^T a a^T \Phi \phi(x_n) - 2 \Phi^T a \phi(x_n) + t_n^2\right) + \frac{1}{2} a^T \Phi \Phi^T a \\
    & = \frac{1}{2} \sum_{n=1}^{N} \left(\phi(x_n)^T a^T \Phi \Phi^T a \phi(x_n) - 2 \Phi^T a \phi(x_n) + t_n^2\right) + \frac{1}{2} a^T \Phi \Phi^T a \\
    & = \frac{1}{2} a^T \Phi\Phi^T \Phi\Phi^T a - a^T \Phi\Phi^T t + \frac{1}{2}t^T t + \frac{\lambda}{2} a^T \Phi\Phi^T a
\end{align}
where $t = (t_1, t_2, ..., t_N)^T$. For further convenience, we define $K_{mn} = \phi(x_m)^T \phi(x_n) = k(x_m, x_n)$ the Gram kernel, then $K = (K_{mn})$ is the Gram matrix with $K = \Phi\Phi^T$.

Following the notation of the Gram matrix, we can rewrite $J(a)$ as

$$
    J(a) = \frac{1}{2} a^T K K a - a^T K t + \frac{1}{2}t^T t + \frac{\lambda}{2} a^T K a
$$

Again, let take the partial derivative of $J(a)$ w.r.t $a$, we have:

$$
    \frac{\delta}{\delta a} J(a) = KKa - Kt + \lambda Ka
$$

\begin{align}
    &                   & \frac{\delta}{\delta a} J(a) = 0 \\
    & \Leftrightarrow   & a(K + \lambda I_N) = t \\
    & \Leftrightarrow   & a = (K + \lambda I_N)^{-1} t \\
\end{align}

Return to the linear regression model, by replacing $a = (K + \lambda I_N)^{-1} t$, we get the predicted value for the new input $x$ as:

$$
    y = w^T \phi(x) = a^T\Phi \phi(x) = ((K + \lambda I_N)^{-1}t)^T k(x) = k(x)^T (K + \lambda I_N)^{-1}t
$$
where we defined $k(x) = (k(x_1, x), k(x_2, x), ..., k(x_N, x))^T$.

The representation above of the linear regression model is called the **dial representation**. By this approach, we obtain a linear regression method base soly on the target variables $(t_n)_n$ and the input variables $(x_n)_n$ without the need of the parameter $w$. From that on, we can construct a regression model to describe **infinite-dimensional data**.

Nevertheless, as we can see in the above formula, whenever we want to approximate an outcome for a new data point $x$, we need to determine the kernel function of this data point $x$ with all training data points $(x_n)_n$. Support Vector Machine (SVM) is a typically kernel method which was introduced to alleviate this problem.

# Support Vector Machine

## Formation

Starting from binary classification using the linear regression model

$$
    y = f(x) = w^T x + b
$$
where $x \in \mathbb{R}^d, y \in \mathbb{R}$.

We have already known that the set $D = \{ x | f(x) = 0 \}$ is a hyperplan determine by vector $w$ in the vector space $\mathbb{R}^{d+1}$ where $dim(D) = d$. This hyperplan $D$ divide the space of $y$ into three subspace: those making $y = 0$, those making $y > 0$ and those making $y < 0$. From that on, we can define the label given the value of $y$: let define $\{-1, 1\}$ the two class of the problem, then $x$ belong to class $1$ if $y > 0$ and class $-1$ otherwise.

If we can determine a hyperplan that completely separates the data space into two subspace, then that dataset is called **linearly separable**. The hyperplan that separates such dataset is called the **decision boundary**.

Before diving into deeper technical steps of determining the best decision boundary, let take an overlook on the hpyerplan in the view of geometry. Considering the following hyperplan $(D)$ determined as:

$$
    y = w^T x + b
$$
In this case, if $x_A$ and $x_B$ are both in this hyperplan $(D)$, then we have

$$
    w^T (x_A - x_B) = 0 \Leftrightarrow w \perp (D)
$$
We can induce this statement as $x_A - x_B$ results in a vector belonging to the hyperplan $(D)$, and the scalar product between $w$ and $x_A - x_B$ results in $0$ which indicates their perpendicular.

Consider a particular $x \in \mathbb{R}^{d}$ and their projected image on $(D)$ $x_\perp \in (D)$. We have

$$
    x = x_\perp + r \frac{w}{||w||}
$$
where $r \in \mathbb{R}$ is any scalar. Multiplying both sides of the formula by $w^T$ and add $b$ to yield

$$
    w^T x + b = (w^T x_\perp + b) + r \frac{w^t w}{||w||} \Leftrightarrow y = r \frac{||w||^2}{||w||} \Rightarrow r = \frac{y}{||w||}
$$
As $\frac{w}{||w||}$ is an unit vector, then $|r|$ is interpreted as the distance between $x$ to the hyperplan $(D)$.

From the definition of the binary classification task above, we need to find $(D): y = w^T x + b$ in order that $\forall x_n, a_n (w^T x_n + b) > 0$ where $a_n \in \{-1, 1 \}$. This implies 

$$
    \frac{a_n (w^T x_n + b)}{||w||}
$$ 
is the distance between $x_n$ and $(D)$.

It is worth noting that on a linearly separable dataset, there can be many possible decision boundaries. A naive problem we have to take into account is which boundary is the best one. In other words, we have to specify what does **best** mean in this scenario. First, let recall the definition of the distance betweeen a point and a set of point. Let $A$ is a set of points and $\alpha$ is any particular point. The distance between $A$ and $\alpha$ is determined as:

$$
    d(A, \alpha) = \underset{x \in A}{min} (d(x, \alpha))
$$
where $d(x, y)$ indicates the distance between two points $x$ and $y$. From that on, the distance between two sets of points $A$ and $B$ is defined as

$$
    d(A, B) = \underset{(x, y) \in A \times B}{min} (d(x, y))
$$

Given a decision boundary $(D)$, we know that its distance to a set of data points $x_n \in S_{-1}$ classified as $-1$ is 
$$
    d_{-1} = \underset{\alpha \in (D), x_n \in S_{-1}}{min}(d(\alpha, x_n))
$$

and similarly for $S_1$ as the set of all points classified as $1$:

$$
    d_1 = \underset{\alpha \in (D), x_n \in S_{1}}{min}(d(\alpha, x_n))
$$

We define the **margin** of a decision boundary is the total distance $d_{-1} + d_1$. From that on, we need to find the parameters $(w, b)$ in order to have the maximum margin:

$$
    \underset{w, b}{argmax} \left( \frac{1}{||w||} \underset{x_n}{min} (w^T x_n + b) \right)
$$

However, directly optimizing to find the optimal $w$ and $b$ is complicated. To simplify the optimization process, we can convert this task to an equivalent task but easier to solve. Considering the distance between $x_n$ and $(D)$, let define a scaling function $f_\kappa: x \mapsto \kappa x$ ($\kappa$ is the scaling factor) on $w$ and $b$, we have:

$$
    \frac{a_n(f_\kappa(w)^T x_n + f_\kappa(b))}{||f_\kappa(w)||} = \frac{a_n (w^T x_n + b)}{||w||}
$$
which means scaling $w$ and $b$ by a scaling factor $\kappa$ the distance between $x_n$ and $(D)$ does not change. To this end, we can deploy a scaling factors $\kappa$ to set 

$$
    a_n (w^T x_n + b) = 1
    
$$
for points that are closest to the decision boundary. Therefore, the distance between $x_n$ and $(D)$ is specified by

$$
    d(x_n, (D)) = \underset{x_n}{min} (a_n (w^T x_n + b)) = 1
$$
Then our objective becomes

$$
    \underset{w, b}{argmax(\frac{1}{||w||})}
$$
which is equivalent to

$$
    \underset{w, b}{argmin \frac{1}{2} w^T w}
$$
that satisfies $a_n (w^T x_n + b) \ge 1$.

## Dual Representation

In order to solve this constrained optimization problem, we introduce the Lagrange multiplier $\lambda_n \ge 0$ for each constraint $a_n(w^\phi(x_n) + b) \ge 1$, which gives the following Lagrange function

$$
    J(w, \lambda, b) = \frac{1}{2} w^t w - \sum_{n=1}^N \lambda_n \left[ a_n (w^t \phi(x_n) + b) - 1 \right]
$$
where $\lambda = (\lambda_1, \lambda_2, ..., \lambda_N)^T$.

Setting the partial derivative of $L(w, \lambda, b)$ w.r.t $w$ and $b$ we have

$$
    \frac{\delta L(w, \lambda, b)}{\delta w} = w - \sum_{n=1}^N \lambda_n a_n \phi(x_n)
$$

$$
    \frac{\delta L(w, \lambda, b)}{\delta b} = -\sum_{n=1}^N \lambda_n a_n 
$$

Setting these partial derivatives to $0$ we have

$$
    w = \sum_{n=1}^N \lambda_n a_n \phi(x_n)
$$

and

$$
    \sum_{n=1}^N \lambda_n a_n = 0
$$

Replacing these terms into $L(w, \lambda, b)$ we have

$$
\begin{align}
    J(\lambda, w) & = \frac{1}{2} \sum_{n=1}^N \sum_{m=1}^N \lambda_n \lambda_m a_n a_m k(x_n, x_m) - \sum_{n=1}^N \lambda_n a_n (w^T \phi(x_n ) + b) - \sum_{n=1}^N \lambda_n \\
    & = \frac{1}{2} \sum_{n=1}^N \sum_{m=1}^N \lambda_n \lambda_m a_n a_m k(x_n, x_m) - \sum_{n=1}^N \lambda_n a_n - \sum_{n=1}^N \lambda_n \\
    & = \frac{1}{2} \sum_{n=1}^N \sum_{m=1}^N \lambda_n \lambda_m a_n a_m k(x_n, x_m) - \sum_{n=1}^N \lambda_n
\end{align}
$$
with $k(x_m, x_n) = \phi(x_m)^T\phi(x_n)$. Now we can obtain its dual representation as

$$
    \tilde{L}(\lambda) = \sum_{n=1}^N \lambda_n - \frac{1}{2} \sum_{n=1}^N \sum_{m=1}^N \lambda_n \lambda_m a_n a_m k(x_n, x_m)
$$

and find its minimum solutions that subject to constraints 

\begin{align}
    \lambda_n & \ge& 0 \\
    \sum_{n=1}^N \lambda_n a_n & = & 0
\end{align}
In addition, thanks to the quadratic form of the kernel $k(x_n, x_m) = \phi(x_n)^T \phi(x_m)$, $\tilde{L}(\lambda)$ is bounded below, which gives raise to a well-defined optimization problem.

In order to classify new data points using the trained model, we evaluate the sign of 

$$
    y(x) = w^T \phi(x) + b = \sum_{n=1}^N \lambda_n a_n k(x_n, x) + b \quad (*)
$$
together with the above-indicated constraints

\begin{align}
    \lambda_n & \ge & 0 \\
    a_n y(x_n) - 1 & \ge & 0 \\
    \lambda_n (a_n y(x_n) - 1) & = & 0
\end{align}

From this on, we can see that either $\lambda_n = 0$ or $a_n y(x_n) - 1 = 0$. Any points for which $\lambda_n = 0$ do not appear in the sum $(*)$ hence they play no roles in making decision for new data points. The remaining ones for which $a_n y(x_n) = 1 $ are call **support vectors**, which is the reason for the name **Support Vector Machine**.

## Overlapping class distribution

The aforementioned SVM deploys fix margins. Such margins are called **hard margins**. In practice, there are cases where some training points are missed classified. These points introduces unstable environments where SVM might fail to determine the decision boundary. We now modify this approach to allow these points lie in the wrong side of the margins, but with a penalty that increases with the distance of them to those margins. To this end, we can provide a slack variable $\xi_n = 0$ where $n \in \{1, 2, ..., N\}$. These slack variables $\xi_n = 0$ if the data points $x_n$ are on or inside the correct margins, while $\xi_n = 1$ if the data points are on the decision boundary $y(x_n) = 0$, and points lying on the wrong side of boundary will have $\xi_n > 1$. From that on, we can define this slack variable as $\xi_n = | a_n - y(x_n) |$. This is described as relaxing the hard constraint of the margin, hence results in the **soft margins**. Note that although we relax the hard margins for allowing wrong-side data points, soft-margin SVM is still sensitive to outlier because the penalty for missclassification increases linearly with $\xi_n$. The classification constraints are then replaced with

$$
    a_n y(x_n) \ge 1 - \xi_n
$$

With this new approach, our goal is to maximize margins while softly penalize the points on the wrong sides, which mean we need to minimize the following objective function:

$$
    \frac{1}{2} \parallel w \parallel^2 + C \sum_{n=1}^N \xi_n
$$
the hyperparameter $C > 0$ controls the trade-off between the slack variable $\xi_n$ and the margin. To see this trade-off, recall that our target is to maximize $\frac{1}{\parallel w \parallel}$, which means minimize $\frac{1}{2} \parallel w \parallel$. Smaller $\parallel w \parallel$ implies wider margins, hence the SVM try its best to classify all of the given points. Given the hyperparameter $C > 0$, higher $C $, smaller $\parallel w \parallel$ and vice versa. In the limit of $C \rightarrow \infty$, the soft-margin mechanism shares the same color as the hard-margin approach.

Now we redefine the Lagrange function for optimizing the objective $\frac{1}{2}\parallel w \parallel^2 + C \sum_{n=1}^N \xi_n$ w.r.t the constraints $a_n y(x_n) \ge 1 - \xi_n$ and $\xi_n > 0$ as:

$$
    L(\lambda, w, b) = \frac{1}{2}\parallel w \parallel^2 + C \sum_{n=1}^N \xi_n - \sum_{n=1}^N \lambda_n (a_n y(x_n) - 1 + \xi_n) - \sum_{n=1}^N \mu_n \xi_n
$$
where $\lambda_n \ge 0$ and $\mu_n \ge 0$ are Lagrange multipliers (or more precisely, the KKT multipliers). According to the KKT theory, the follow conditions are hold:

$$
\begin{align}
    \lambda_n & \ge & 0 \\
    \mu_n & \ge &  0 \\
    \xi_n & \ge & 0 \\
    a_n y(x_n) - 1 + \xi_n & \ge & 0 \\
    \lambda_n (a_n y(x_n) - 1 + \xi_n) & = & 0 \\
    \xi_n \mu_n & = & 0 
\end{align}
$$

Taking the derivatives of $L(\lambda, w, b)$ w.r.t $w$, $b$, and $\xi_n$, we have:

$$
    \frac{\delta L}{\delta w} = 0 \Leftrightarrow w = \sum_{n=1}^N \lambda_n a_n \phi(x_n)
$$

$$
    \frac{\delta L}{\delta b} = 0 \Leftrightarrow \sum_{n=1}^N \lambda_n a_n = 0
$$

$$
    \frac{\delta L}{\delta \xi_n} = 0 \Leftrightarrow \lambda_n = C - \mu_n
$$

Using these results to obtain the dual form of $L(\lambda, w, b)$ we have:

$$
    \tilde{L}(\lambda) = \sum_{n=1}^N a_n - \frac{1}{2} \sum_{n=1}^N \sum_{m=1}^N \lambda_n \lambda_n a_n a_m k(x_n, x_m)
$$
Substituting $w = \sum_{n=1}^N \lambda_n a_n \phi(x_n)$ into $y(x) = w^T \phi(x) + b$ we will obtain the same result as $(*)$.


## Hinge loss for SVM

We have seen that for data points relies on or in the correct side then $y_n a_n = 1$ hence $\xi_n = 0$, and for the remaining data points we have $\xi_n = 1 - y_n a_n$. The objective function 

$$
    \frac{1}{2}\parallel w \parallel^2 + C \sum_{n=1}^N \xi_n 
$$
can be rewritten as

$$
    \frac{1}{2} \parallel w \parallel^2 + C \sum_{n=1}^N \parallel 1 - y_n a_n \parallel
$$
This objective function is named **Hinge error function** or **Hinge loss**.

Another way to express this loss is

$$
    \mathcal{l}_{Hinge} = \frac{1}{2} \parallel w \parallel^2 + C \sum_{n=1}^N max(0, 1 - y_n a_n)
$$

The term $l_i = max(0, 1 - y_n a_n)$ is not diffentiable, but subdifferentiable. We can consider two cases for determining its derivative as follow:

- Case 1: $y_n a_n \ge 1:$

In this case, $l_i = 0$, hence we have $\frac{\delta l_i}{\delta w} = 0$.

- Case 2: $y_n a_n < 1:$

In this case, $l_i = 1 - y_n a_n$, then:

$$
    \frac{\delta l_i}{\delta w} = \frac{\delta }{\delta w} \left[ 1 - a_n (w^T x_n + b) \right] = -a_n x_n
$$
From this on, we have:

$$
    \nabla_w \mathcal{l}_{Hinge} = w + C \sum_{i: y_n a_n < 1 } (-a_n x_n)
$$

Given the Hinge loss, we can use gradient descent method to train the soft-margin SVM model.
